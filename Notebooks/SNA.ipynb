```
consumo_dir = '/content/drive/MyDrive/0AccTeam/datasets_consumo/'
```

```
edges = pd.read_csv(consumo_dir + "reseña_sin_comentario_consumo.csv")
```

```
negocio_nodes = pd.read_csv(consumo_dir + "negocio_consumo.csv", encoding="charmap")
usuario_nodes = pd.read_csv("/content/drive/MyDrive/0AccTeam/usuario.csv")
```

```
negocio_nodes = negocio_nodes[negocio_nodes["cantidad_reseÃ±as"]>10]
```

```
negocio_nodes = negocio_nodes[["negocio_id","nombre","estrellas", "latitud", "longitud" ]]
negocio_nodes = negocio_nodes.rename(columns = {"negocio_id": "id", "nombre":"label"})
```

```
usuario_nodes = usuario_nodes[["usuario_id", "nombre", "cantidad_fans"]]
usuario_nodes = usuario_nodes.rename(columns = {"usuario_id": "id", "nombre":"label"})
```

```
usuario_nodes = usuario_nodes[usuario_nodes.cantidad_fans > 2]
```

```
edges = edges[edges.usuario_id.isin(usuario_nodes.id)]
edges = edges[edges.negocio_id.isin(negocio_nodes.id)]
```

```
usuario_nodes["id_number"] = range(0,len(usuario_nodes.index)) 
negocio_nodes["id_number"] = range(len(usuario_nodes.index), len(usuario_nodes.index) + len(negocio_nodes.index))
```

```
usuario_ids = usuario_nodes[["id","label", "id_number"]].set_index("id")
negocio_ids = negocio_nodes[["id","label", "id_number"]].set_index("id")
```

```
edges = edges[["usuario_id", "negocio_id", "estrellas"]].rename(columns = {"usuario_id":"source", "negocio_id":"target"})
```

```
edges.source = edges.source.apply(lambda s: usuario_ids.loc[s].id_number)
edges.target = edges.target.apply(lambda s: negocio_ids.loc[s].id_number)
```

```
edges.to_csv(consumo_dir + "SNA/edges.csv", index=False)
```

```
# 10 para usuario, 100 para negocio
usuario_nodes["type"] = 10
negocio_nodes["type"] = 100
```

```
usuario_nodes = usuario_nodes.rename(columns = {"id":"usuario_id", "id_number": "id"})
negocio_nodes = negocio_nodes.rename(columns = {"id":"negocio_id", "id_number": "id"})
```

```
usuario_nodes.to_csv(consumo_dir + "SNA/usuarios_nodes.csv", index=False)
negocio_nodes.to_csv(consumo_dir + "SNA/negocio_nodes.csv", index=False)
```
