```python
def insert_row(idx, df, df_insert):
    dfA = df.iloc[:idx, ]
    dfB = df.iloc[idx:, ]

    df = dfA.append(df_insert).append(dfB).reset_index(drop = True)

    return df
```

## Por Estado

### Por numero de Negocio

```python
estado_group = data.groupby('states').count()
```

```python
estado_group#.estado.unique()
```

```python
estado_group.describe()
```

```python
estado_group2 = estado_group.sort_values('nombre',ascending=False)
estado_group3 = estado_group2[estado_group2['negocio_id']>3]
estado_group3
```

```python
estado_group3.negocio_id.mean()
```

```python
from bokeh.plotting import figure, show

p = figure(y_range = estado_group3.index.values,
           plot_width = 800,
           plot_height = 600,
           title = 'Business by State',
           x_axis_label = 'Count of Business',
           tools = "pan,box_select,zoom_in,zoom_out,save")
p.hbar(y = estado_group3.index.values,
       right = estado_group3.nombre,
       left = estado_group3.negocio_id.mean(),
       height = 0.4,
       color = 'orange',
       fill_alpha = 0.5)
show(p)
```

#### Porcentaje de data por estado

```python
for i, state in enumerate(list(estado_group3.index.values)):
  chunk_ = data[data['states']==state]
  if i==0:
    data_rude = chunk_
  else:
    data_rude.append(chunk_,ignore_index=True)
```

```python
sns.heatmap(data_rude.isnull(), cbar=False)
```

```python
data_rude.shape
```

```python
len(data_rude.columns[8:])
```

```python
data_rude.head()
```

```python
data_rude['full_count'] = data_rude.apply(lambda x: round(((x.count()-8)/49)*100,2), axis=1)
```

```python
data_rude.full_count.unique()
```

```python
data_rude2 = data_rude[data_rude['full_count']>50]
```

```python
missing_data = data_rude2.isna().mean().round(4)*100
variables_ = list(missing_data.sort_values(ascending=False).where(missing_data<30).dropna().index.values)
```

```python
data_rude3 = data_rude2[variables_]
```

```python
data_rude3.isna().mean().round(4)*100
```

```python
heat_map = sns.heatmap(data_rude3.isna(), cmap="YlGnBu",xticklabels=True)
plt.show()
```
