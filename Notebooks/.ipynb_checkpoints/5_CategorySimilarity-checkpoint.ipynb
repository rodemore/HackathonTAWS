```python
embedding = load_embedding(EMBEDDING_PATH)
```

```python
def comment_to_vector(comment, prefix=False):
    sentence = np.zeros([embedding.dims])
    num_words = 0

    for word in comment:
        try:
            if prefix:
                sentence += embedding.word_to_vector(prefix+'/'+word)
            else:
                sentence += embedding.word_to_vector(word)
            num_words += 1
        except KeyError as r:
            continue

    if num_words > 0:
        sentence = sentence / num_words
    else:
        logging.warning(
            'Could not calculate the sentence embedding fot the sentence ' + str(comment))

    return sentence    
    #return self.pad_sentence(sentence, embedding.dims)
```

```python
def get_distances(text, cat_vector):
    normalized_text = normalize_comments(text)
    #dimension de vectores de word embedding
    #print(comment_to_vector(text))
    vector = comment_to_vector(normalized_text)
    #print(vector.size)
    return spatial.distance.cosine(cat_vector, vector)
    #return distance.euclidean(cat_vector,vector )
```
